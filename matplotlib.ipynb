```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,1,8,7])
```

```python
plt.plot(x,y)
plt.show()
```

Marker|Line|Color

```python
plt.plot(x,y,'x--r')
plt.show()
```

```python
plt.plot(x,y,'o:r', ms=20, mec='b', mfc='#3cb371')
plt.show()
```

```python
plt.plot(x,y, color='r', ls=':')
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,1,8,7])
z = np.array([13,15,20,17,12])

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,1,8,7])
z = np.array([13,15,20,17,12])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```

```python
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
plt.grid(axis='y', color='#3cb371', linestyle='-.', linewidth=2)
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('X Axis')
plt.ylabel('y Axis')
plt.plot(x,y)
plt.show()
```

```python
plt.title('Chart')
plt.xlabel('X Axis')
plt.ylabel('y Axis')
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y, width=0.5, color='r')
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```
